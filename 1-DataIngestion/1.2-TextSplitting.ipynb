How to recursively split text by characters - RecursiveCharacter Text Splitter

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)

final_documents = text_splitter.split_documents(pdf_documents)
# final_documents
print(final_documents[0])
print(final_documents[1])
```

How to recursively split text by characters - Character Text Splitter

```python
from langchain_community.document_loaders import TextLoader

loader = TextLoader(file_path="speech.txt")

text_documents = loader.load()


from langchain_text_splitters import CharacterTextSplitter
text_splitter = CharacterTextSplitter(separator="\n", chunk_size=50, chunk_overlap=10)
text_splitter.split_documents(text_documents)
```

How to recursively split text by characters - HTML Header Text Splitter

```python
from langchain_text_splitters import HTMLHeaderTextSplitter

html_string = """
<html>
  <head><title>Sample Page</title></head>
  <body>
    <h1>Main Title</h1>
    <p>This is a paragraph under the main title.</p>
    <h2>Subsection</h2>
    <p>Details about the subsection.</p>
  </body>
</html>
"""

# Correct usage: list of (header_tag, header_name) tuples
headers_to_split_on = [
    ("h1", "Main Title"),
    ("h2", "Subsection")
]

splitter = HTMLHeaderTextSplitter(headers_to_split_on=headers_to_split_on)
splits = splitter.split_text(html_string)

for i, split in enumerate(splits):
    print(f"Chunk {i+1}:\n{split}\n")
```

```python
import requests

url = "https://en.wikipedia.org/wiki/Artificial_intelligence"
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)"}
response = requests.get(url, headers=headers)
html_string = response.text

from langchain_text_splitters import HTMLHeaderTextSplitter

# You may want to split on generic headers, since Wikipedia's h1/h2 text varies
headers_to_split_on = [
    ("h1", ""),  # All h1 tags
    ("h2", ""),  # All h2 tags
    ("h3", ""),  # All h3 tags
]

splitter = HTMLHeaderTextSplitter(headers_to_split_on=headers_to_split_on)
splits = splitter.split_text(html_string)

splits
```

```python
import json
import requests

json_data = requests.get("https://api.spacexdata.com/v4/launches/latest").json()

from langchain_text_splitters import RecursiveJsonSplitter
splitter = RecursiveJsonSplitter(max_chunk_size=300)
json_chunks = splitter.split_json(json_data)

json_chunks
for chunk in json_chunks[:3]:
    print(chunk)
```
