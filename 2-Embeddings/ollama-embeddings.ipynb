```python
from langchain_community.embeddings import OllamaEmbeddings

#by default it uses the model "llama2"
embeddings = OllamaEmbeddings(model="gemma:2b")
embeddings
```

```python
r1 = embeddings.embed_documents(["This is a sample document", "This is another sample document"])
r1[0]
```

```python
r2 = embeddings.embed_query("This is a sample document")
r2
```
